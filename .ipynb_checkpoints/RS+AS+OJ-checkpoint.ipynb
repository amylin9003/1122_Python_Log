```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import hashlib as hb
```

```python
def md5(i):
    hl = hb.md5()
    hl.update(str(i).encode(encoding='utf-8'))   
    # print(str(i)) # 加密前
    # print(hl.hexdigest()) # 加密後
    return hl.hexdigest()
```

# RS

```python
# 每一類別題目總數
rs_pd = pd.read_csv('../data/1111_RS_log.csv')
rs_pd
```

```python
submit_log = rs_pd[rs_pd['operation_name']=='SUBMIT']
submit_log = submit_log.dropna()

keep_list = ['contents_name', 'username', 'sentence_id', 'result']
submit_log = submit_log[keep_list]
submit_log
```

```python
'''
    計算每一類別分別有多少題目

'''
userid_list = list(set(submit_log['username'])) # unique username
contents_name_list = list(set(submit_log['contents_name'])) # 題目類別
contents_name_list.sort()
sentence_id_list = list(set(submit_log['sentence_id'])) # 題目編號

num_contents = [0] * len( contents_name_list ) # 每一類別有幾題
isVisited = [False] * len( sentence_id_list ) # 此題號是否已被計算過


for contents_name, sentence_id in zip( submit_log['contents_name'], submit_log['sentence_id'] ) :
    if ( isVisited[sentence_id_list.index(sentence_id)] ) : 
        continue
    num_contents[contents_name_list.index(contents_name)] += 1
    isVisited[sentence_id_list.index(sentence_id)] = True

num_contents
            
```

```python
'''
    計算每位同學在每個章節分別submit的對與錯次數
    userid :
    submit_wrong : []
    submit_accept : []
'''
userid_list = list(set(submit_log['username'])) # unique username

contents_name_list = list(set(submit_log['contents_name'])) # 題目類別
contents_name_list.sort()

sentence_id_list = list(set(submit_log['sentence_id'])) # 題目編號


user_submit = [] * len(userid_list)# [submit_wrong,submit_accept] * len(userid_list)

for user in userid_list :
    user_dict = { 'userid' : user, 'submit_wrong' : [], 'submit_accept' : [] }
    for contents_name in contents_name_list :
        log_list = submit_log[submit_log['username'] == user][submit_log['contents_name'] == contents_name]
        submit_wrong = 0
        submit_accept = 0
        for log in log_list['result'] :
            if ( log == 1.0 ) :
                submit_accept += 1
            else :
                submit_wrong += 1
        user_dict['submit_wrong'].append(submit_wrong)
        user_dict['submit_accept'].append(submit_accept)
    user_submit.append(user_dict)
user_submit
```

```python
'''
    將dict轉成dataframe
'''

# 先取出'每章節'每個人'平均'的accept與wrong題數
userid = []
course = [[] for i in range(55)]

rs_accept = []
rs_wrong = []

for i in range(len(user_submit)):
    userid.append(user_submit[i]['userid'])
    for j in range(18):
        if j % 2 == 0:
            course[i].append( round(user_submit[i]['submit_accept'][j//2] / num_contents[j//2], 2) )
        else:
            course[i].append( round(user_submit[i]['submit_wrong'][j//2] / num_contents[j//2], 2) )
            
            
    rs_accept.append(sum(user_submit[i]['submit_accept']))
    rs_wrong.append(sum(user_submit[i]['submit_wrong']))
    

# 取出章節名稱
contents_name_list = list(set(submit_log['contents_name'])) # 題目類別
contents_name_list.sort()
# print(contents_name_list)

# 定義每個columns名稱
col_name_list = []
for col_name in contents_name_list :
    if col_name[:2] == 'C2':
        col_name_list.append('rs.'+col_name[:4]+'_ac')
        col_name_list.append('rs.'+col_name[:4]+'_wr')
    else  :
        col_name_list.append('rs.'+col_name[:2]+'_ac')
        col_name_list.append('rs.'+col_name[:2]+'_wr')
# col_name_list



# 包裝成DataFrame
rs_pd = pd.DataFrame(course, columns = col_name_list)
rs_pd["username"] = userid
rs_pd["userid"] = rs_pd["username"].apply(md5)
rs_pd = rs_pd.reindex(columns = ['userid'] + col_name_list )
rs_pd['rs_accept'] = rs_accept
rs_pd['rs_wrong'] = rs_wrong
rs_pd
```

```python
as_pd = pd.read_csv('../data/1111_AS_log.csv')
as_log = as_pd[['username',  'contents_name', 'status', 'result']]
as_log
```

```python
as_content_name = list(set(as_pd['contents_name']))
as_content_name.sort()
as_content_name
```

```python
'''
    計算每位同學在每個章節Accept、Countinue、Reject次數
    userid :
    submit_wrong : []
    submit_accept : []
'''
userid_list = list(set(as_log['username'])) # unique username
contents_name_list = list(set(as_log['contents_name'])) # 題目類別
contents_name_list.sort()


user_submit = [] * len(userid_list)# [submit_wrong,submit_accept] * len(userid_list)

for user in userid_list :
    user_dict = { 'userid' : user, 'Accept' : [], 'Continue' : [], 'Reject' : [] }
    for contents_name in contents_name_list :
        log_list = as_log[as_log['username'] == user][as_log['contents_name'] == contents_name]
        # print(log_list)
        as_accept = 0
        as_continue = 0
        as_reject = 0
        for log in log_list['status'] :
            if ( log == 'Accept' ) :
                as_accept += 1
            elif ( log == 'Countinue' ) :
                as_continue += 1
            else :
                as_reject += 1
                
        user_dict['Accept'].append(as_accept)
        user_dict['Continue'].append(as_continue)
        user_dict['Reject'].append(as_reject)
    user_submit.append(user_dict)
user_submit
```

```python
'''
    將dict轉成dataframe
'''

# 先取出'每章節'每個人'平均'的accept與wrong題數
userid = []
course = [[] for i in range(55)]

as_accept = []
as_continue = []
as_reject = []

for i in range(len(user_submit)):
    userid.append(user_submit[i]['userid'])
    for j in range(24):
        
        if j % 3 == 0:
            course[i].append( round(user_submit[i]['Accept'][j//3] ))
        elif j % 3 == 1:
            course[i].append( round(user_submit[i]['Continue'][j//3] ))                  
        else:
            course[i].append( round(user_submit[i]['Reject'][j//3] ))
    as_accept.append(sum(user_submit[i]['Accept']))
    as_continue.append(sum(user_submit[i]['Continue']))
    as_reject.append(sum(user_submit[i]['Reject']))
            
# 取出章節名稱
contents_name_list = list(set(as_log['contents_name'])) # 題目類別
contents_name_list.sort()
# print(contents_name_list)

# 定義每個columns名稱
col_name_list = []
for col_name in contents_name_list :
    col_name_list.append('as.'+col_name[:2]+'_ac')
    col_name_list.append('as.'+col_name[:2]+'_ct')
    col_name_list.append('as.'+col_name[:2]+'_rj')
# col_name_list

# 包裝成DataFrame
as_pd = pd.DataFrame(course, columns = col_name_list)
as_pd["username"] = userid
as_pd["userid"] = as_pd["username"].apply(md5)
as_pd = as_pd.reindex(columns = ['userid'] + col_name_list )
as_pd['as_accept'] = as_accept
as_pd['as_continue'] = as_continue
as_pd['as_reject'] = as_reject
as_pd
```

# Kmeans

```python
merge_data = merge_pd.drop(['userid', 'score', 'class'], axis=1)
merge_score = merge_pd['score']
```

```python
print('[全做Kmeans]')
X = merge_data
# KMeans 演算法
kmeans_fit = KMeans(n_clusters = 2,random_state=0).fit(X)

# 印出分群結果
cluster_labels = kmeans_fit.labels_ # 或是 =  KM.predict(X) #
print("分群結果：")
print(cluster_labels)
print("---")

# 印出score
print("真實分數：")
print(merge_score.values)

u_labels = np.unique(cluster_labels)
for feature in merge_data.columns: 
    plt.figure()
    for i in u_labels:
        plt.title(feature)
        plt.scatter(X[feature][X[cluster_labels==i].index] , merge_score[merge_score[cluster_labels==i].index], label = i)
    plt.legend(loc='lower right')
    results_path = '../plot/KMeans/RS_AS_OJ/' + feature + '.png'
    # plt.savefig(results_path, dpi='figure', bbox_inches='tight')
```
