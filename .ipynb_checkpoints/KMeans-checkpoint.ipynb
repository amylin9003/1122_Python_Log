```python
from sklearn.cluster import KMeans
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import hashlib as hb
```

# BookRoll + VisCode + OJ Problem

```python
br_pd = pd.read_csv('../data/br.csv')
br_pd = br_pd.drop(['Unnamed: 0', 'ADD_RECOMMENDATION', 'REGIST CONTENTS', 'CLOSE_RECOMMENDATION', 'OPEN_RECOMMENDATION'], axis =1)
br_pd = br_pd[br_pd['class']=='h']
br_pd = br_pd.drop('class', axis=1)
```

```python
viscode_pd = pd.read_csv('../data/viscode.csv')
viscode_pd = viscode_pd.drop(['Unnamed: 0' ], axis=1)
viscode_pd = viscode_pd[viscode_pd['class']=='h']
viscode_pd =  viscode_pd.drop(['class', 'score'], axis=1)
```

```python
oj_problem = pd.read_csv('../data/1111_OJ_log.csv')
oj_list = []
for i in range(len(oj_problem)) :
    log = oj_problem.iloc[i]
    # print(log.problem )
    if 'P' in log.problem:
        oj_list.append(log)
oj_prob = pd.DataFrame(oj_list)
oj_prob
```

```python
oj_status_grouped = oj_prob.groupby(['username', 'status']).size().reset_index()
oj_status_grouped = oj_status_grouped.pivot_table(0, ['username'], 'status').fillna(0).astype('int')
oj_status_grouped
```

```python
def md5(i):
    hl = hb.md5()
    hl.update(str(i).encode(encoding='utf-8'))   
    # print(str(i)) # 加密前
    # print(hl.hexdigest()) # 加密後
    return hl.hexdigest()
```

```python
user_list = pd.DataFrame( oj_status_grouped.index)
user_list = user_list
```

```python
test = user_list['username'].apply(md5)
oj_status_grouped['userid'] = list(test)
oj_status_grouped
```

```python
merge_pd = pd.merge(br_pd, viscode_pd, on='userid', how = 'inner')
merge_pd = pd.merge(merge_pd, oj_status_grouped, on='userid', how = 'inner')
merge_pd
```

```python
# merge_pd.to_csv('Learning System.csv')
```

```python
merge_pd = merge_pd.drop(['userid', 'Viscode-spent_time'], axis=1)
merge_pd
```

```python
keep_list = ['ADD MARKER', 'DELETE_MEMO', 'LINK_CLICK', 'code_copy', 'code_execution', 'code_paste', 'code_speed' , 
             'notebook_open', 'KeyError', 'KeyboardInterrupt', 'codeLength', 'Viscode-login_times', 'Viscode-execute_times',
             'Viscode-open_file_times', 'Viscode-success_run', 'Viscode-error_types', 
             'oj.Accept']
```

```python
merge_data = merge_pd.drop('score', axis=1)
merge_score = merge_pd['score']
```

```python
print('[全做Kmeans]')
X = merge_data
# KMeans 演算法
kmeans_fit = KMeans(n_clusters = 2,random_state=0).fit(X)

# 印出分群結果
cluster_labels = kmeans_fit.labels_ # 或是 =  KM.predict(X) #
print("分群結果：")
print(cluster_labels)
print("---")

# 印出score
print("真實分數：")
print(merge_score.values)

u_labels = np.unique(cluster_labels)
for feature in keep_list: 
    plt.figure()
    for i in u_labels:
        plt.title(feature)
        plt.scatter(X[feature][X[cluster_labels==i].index] , merge_score[merge_score[cluster_labels==i].index], label = i)
    plt.legend(loc='lower right')
    results_path = './plot/KMeans/BR_VC_OJ_drop_score/' + feature + '.png'
    plt.savefig(results_path, dpi='figure', bbox_inches='tight')
```
